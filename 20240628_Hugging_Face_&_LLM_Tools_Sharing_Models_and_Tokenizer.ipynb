##login with local file

每次要輸入 Hugging Face API 密碼太麻煩  
把密碼明碼寫在程式裡不是一個良好的習慣  
找到一個方法可以在第一次使用詢問的方式接收密碼  
同時寫入到 Google Drive 私人的檔案位置  
之後就可以使用這個私人檔案內的密碼做 login

````python
import os
from google.colab import drive
from getpass import getpass
from huggingface_hub import login


print("This program will mount your Google Drive and check if you have a file named hf_token.txt in your Google Drive. \n\
If the file exists, the program will use the Hugging Face access token found there to log in. \n\
If the file cannot be found, the program will prompt you for your access token, \n\
save it to your Google Drive, and log in with that token.\n\
==========================================")

# Mount Google Drive
drive.mount('/content/drive')

# Define the token file path
token_file_path = '/content/drive/MyDrive/hf_token.txt'

# Function to prompt user for token and save it
def prompt_and_save_token():
    hf_token = getpass("Please enter your Hugging Face access token: ")
    with open(token_file_path, 'w') as token_file:
        token_file.write(hf_token)
    os.environ['HUGGINGFACE_TOKEN'] = hf_token
    os.environ['HF_TOKEN'] = hf_token
    login(token=hf_token)
    print("Token saved to file and environment variable.")


# Check if the token file exists
if os.path.exists(token_file_path):
  with open(token_file_path, 'r') as token_file:
    hf_token = token_file.read().strip()
  os.environ['HUGGINGFACE_TOKEN'] = hf_token
  os.environ['HF_TOKEN'] = hf_token
  login(token=hf_token)
  print("Token loaded from file and set to environment variable.")
else:
    prompt_and_save_token()
````

##notebook_login()

`notebook_login()` 是 Hugging Face Hub 提供的一個簡便方法，用於在 Jupyter Notebook 或 Google Colab 中進行 Hugging Face Hub 的身份驗證。當你運行 `notebook_login()` 時，它會提示你輸入你的 Hugging Face 訪問令牌（token），以便你能夠訪問和操作 Hugging Face Hub 上的資源。

### 具體作用

1. **身份驗證**：`notebook_login()` 會打開一個輸入框，要求你輸入 Hugging Face Hub 的訪問令牌。這個訪問令牌是你在 Hugging Face Hub 上生成的，用於授權你對 Hugging Face Hub 上的儲存庫進行讀寫操作。
2. **設置環境變數**：輸入令牌後，`notebook_login()` 會將該令牌設置為環境變數，讓後續的 Hugging Face Hub 操作自動使用這個令牌進行身份驗證，無需再次手動輸入。
3. **便捷性**：這種方法非常方便，特別是當你在筆記本環境中進行實驗或開發時，無需手動配置複雜的身份驗證步驟。

### 使用方法

你可以在 Jupyter Notebook 或 Google Colab 中運行以下代碼來登錄 Hugging Face Hub：

```python
from huggingface_hub import notebook_login

notebook_login()
```

運行此代碼後，會彈出一個輸入框，要求你輸入 Hugging Face 訪問令牌。你可以在 Hugging Face Hub 的[個人設定頁面](https://huggingface.co/settings/tokens)生成這個訪問令牌。

##HfApi

在 `notebook_login` 中目前並不支持直接傳遞令牌作為參數。但是，你可以使用 Hugging Face Hub 提供的另一種方法來設置環境變數，以自動進行身份驗證。

你可以將 `HUGGINGFACE_TOKEN` 設置為環境變數，這樣 Hugging Face 的各種方法都能自動使用該令牌進行身份驗證。

以下是示例代碼，演示如何設置環境變數並使用 `HfApi` 來登錄：

```python
import os
from huggingface_hub import HfApi

# 將 HUGGINGFACE_TOKEN 設置為環境變數
os.environ['HUGGINGFACE_TOKEN'] = 'your_hf_token'

# 使用 HfApi 登錄
api = HfApi()
api.set_access_token(os.environ['HUGGINGFACE_TOKEN'])

# 或者使用 login 方法
api.login(token=os.environ['HUGGINGFACE_TOKEN'])
```

##Some weights - not used

這個警告信息表示在從指定的檢查點 `camembert-base` 初始化 `CamembertForMaskedLM` 模型時，檢查點中的某些權重沒有被使用。具體來說，以下權重沒有被使用：

- `roberta.pooler.dense.bias`
- `roberta.pooler.dense.weight`

### 原因解釋

這個警告是預期行為的一部分，可能原因如下：

1. **不同任務的模型**：你正在從一個模型檢查點（`camembert-base`）中初始化一個 `CamembertForMaskedLM` 模型，這個檢查點可能是為了另一個任務（如分類任務）訓練的。因此，該檢查點包含了一些在當前任務（掩碼語言模型任務）中不需要的權重。

2. **模型架構不同**：檢查點的模型架構可能與 `CamembertForMaskedLM` 模型的架構略有不同。例如，`camembert-base` 模型可能包括了一些用於分類層的權重（如 `pooler.dense`），而這些權重在掩碼語言模型中不需要使用。

### 是否應該擔心？

這個警告分為兩種情況：

- **預期行為**（如警告所述的情況）：如果你是從另一個任務的檢查點初始化模型，比如從一個預訓練的分類模型初始化掩碼語言模型，這是正常的。這些多餘的權重只是未被使用，對模型的性能和功能沒有影響。

- **非預期行為**：如果你期望這個檢查點與當前任務的模型完全相同（例如從一個掩碼語言模型檢查點初始化另一個掩碼語言模型），那麼這個警告表明你的預期有誤，可能需要檢查檢查點的來源和模型的兼容性。

##upload_file

以下是如何使用 Hugging Face Hub 的 `upload_file` 函數將本地文件上傳到 Hugging Face 儲存庫中的說明和示例代碼：

### 說明

`upload_file` 函數用於將本地文件上傳到 Hugging Face Hub 儲存庫。這對於分享模型的配置文件、權重文件等非常有用。該函數允許指定本地文件路徑、儲存庫內文件的目標路徑以及儲存庫的名稱。

### 參數解釋

- `<path_to_file>`：本地文件的完整路徑。例如，如果文件位於 Google Drive 中，路徑可能是 `/content/drive/MyDrive/config.json`。
- `path_in_repo`：文件在儲存庫中的路徑和名稱。例如，`config.json`。
- `repo_id`：目標儲存庫的 ID，包括命名空間。例如，`ssupinma/dummy-model`。

### 示例代碼

以下代碼展示了如何將位於 Google Drive 中的本地文件 `/content/drive/MyDrive/config.json` 上傳到 Hugging Face 儲存庫中的 `config.json` 文件。

```python
from huggingface_hub import upload_file

# 上傳本地文件到 Hugging Face 儲存庫
upload_file(
    path_or_fileobj="/content/drive/MyDrive/config.json",  # 本地文件路徑
    path_in_repo="config.json",                            # 儲存庫中的目標路徑
    repo_id="ssupinma/dummy-model"                      # 儲存庫 ID
)
```

### 具體步驟

1. **確保已經登錄 Hugging Face Hub**：在上傳文件之前，確保你已經配置好 Hugging Face Hub 的身份驗證（例如使用 `HfApi` 或 `notebook_login` 方法登錄）。
2. **指定文件路徑**：將本地文件的完整路徑 `/content/drive/MyDrive/config.json` 填入 `path_or_fileobj` 參數中。
3. **指定儲存庫和目標路徑**：設置儲存庫 ID 和文件在儲存庫中的路徑。

這樣，文件就會被上傳到指定的 Hugging Face 儲存庫中。
